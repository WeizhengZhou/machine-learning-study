```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Reading data
x_data = pd.read_csv('./data/logistic_x.txt', delimiter='\s+', dtype=np.float64, header=None)
x_data.head()
```

```python
y_data = pd.read_csv('./data/logistic_y.txt', dtype=np.float64, header=None)
```

```python
y_data.head()
```

```python
x1 = x_data[:][0]
x2 = x_data[:][1]
y = y_data
```

```python
positive_y = np.ma.masked_where(y > 0, y)
negative_y = np.ma.masked_where(y < 0, y)
```

```python
plt.scatter(x1, x2, positive_y * 10, marker='o')
# plt.scatter(x1, x2, negative_y * 10, marker='x')
plt.show()
```
